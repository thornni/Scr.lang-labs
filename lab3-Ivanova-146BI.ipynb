# Лабораторная работа №3

Задание: Используя Python и модуль requests и bs4 написать скрипт, извлекающий новости (отдельно заголовок, анотацию, авторов) из веб-страницы новостного агенства (напр. washingtonpost.com). Требуется использовать поиск по дереву html, а не регулярные выражения


## Решение

```python
# Импортируем модули
import requests
from bs4 import BeautifulSoup

sitename = 'https://www.washingtonpost.com/'  
print('Please, wait a minute...')
r = requests.get(sitename)
         
# в эту СТРОКУ запишем содержимое, взятое с исследуемого сайта    
sitecontent = r.text    
    
soup = BeautifulSoup(sitecontent, 'html.parser')

print("\n\n******************** General title is: **********************************") 
print(soup.title.string, "\n\n")

border = "\n\t *  *   *   *   *   *   *   *\n";
mini_soups = list((soup.find_all("div", {"class": "no-skin flex-item flex-stack normal-air text-align-left equalize-height-target"})))

for i in range (0, len(mini_soups)):
    title = mini_soups[i].findAll("a", {"data-pb-field": "web_headline"})
    summary = mini_soups[i].findAll("div", {"data-pb-field": "summary"})
    authors = mini_soups[i].findAll("span", "author vcard")
    
    # авторов бывает несколько для одной статьи
    count_of_authors = len(authors)
    
    if len(title) != 0:
        print("Title:   ", title[0].text)
        if len(summary) != 0:    
            print("\nSummary:   ", summary[0].text)
            
        for k in range(0, count_of_authors):
            if k == 0:
                print("\nAuthors: ")
            print(authors[k].text)
    
        print(border)
```
